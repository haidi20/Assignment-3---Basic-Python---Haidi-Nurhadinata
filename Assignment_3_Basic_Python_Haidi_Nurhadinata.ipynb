4. Buatlah program untuk menghitung luas segitiga dengan dua cara:

      a. Menghitung luas segitiga dengan rumus (alas x tinggi)/2

      b. Menghitung luas segitiga dengan rumus heron


```python
import math

print("1. luas segitiga")
print("2. luas segitiga dengan rumus heron")
menu = input("Pilih menu = ")

def luasSegitiga(alas, tinggi):
  return alas * tinggi / 2

def luasSegitigaHeron(a, b, c):
  s = (a + b + c) * 1/2
  return math.sqrt(s*(s - a) * (s - b) * (s - c))

if(int(menu) == 1):
  alas = input("masukkan nilai alas = ")
  tinggi = input("masukkan nilai tinggi = ")

  result = luasSegitiga(int(alas), int(tinggi))

  print(f"Luas segitiga dengan rumus (alas x tinggi)/2 = {result}")

elif(int(menu) == 2):
  
  a = input("masukkan nilai a = ")
  b = input("masukkan nilai b = ")
  c = input("masukkan nilai c = ")

  luas = luasSegitigaHeron(int(a), int(b), int(c))

  print(f"luas segitiga dengan rumus heron = {luas}")
else:

  print("maaf, tidak ada di pilihan")
```

2. Buatlah program untuk menghitung regresi linier menggunakan metode invers matriks
x = (ATA)-1ATy
dimana:

* x adalah koefisien dari model regresi yang ingin dicari
* A adalah matriks data
* AT adalah transpose dari A
* A-1 adalah invers dari matriks A
* y adalah variabel target

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import requests

csvUrl = "https://raw.githubusercontent.com/hadimaster65555/dataset_for_teaching/main/dataset/linear_regression_simulation_data/linear-regression.csv"
# getData = requests.get(csvUrl)
# getContent = getData.content

def convertCsvToArray(url):
    response = requests.get(url)
    lines = response.text.splitlines() 
    for line in lines[1:]:
        el = [i.strip() for i in line.split(',')]
        yield el 

genData = convertCsvToArray(csvUrl)
listData = list(genData)

A = np.array(listData, dtype=float)
At = np.array(A.transpose(), dtype=float)

x = np.dot(np.linalg.inv(np.dot(At, A)), At)   

result = np.array(x, dtype=float)

plt.plot(x[0], x[1], "o")
print(np.amax(x))
 
# To show the plot
plt.show()
```
