2. Buatlah program yang meminta user untuk mengisi panjang dan lebar dari suatu ruangan. Peserta diminta untuk memberikan output dengan dua satuan yang berbeda:
Luas dalam meter persegi
Luas dalam hektar

```python
inputPanjang = input("masukkan panjang = ")
inputLebar = input("masukkan lebar = ")

def luasPersegi(panjang, lebar):
  return int(panjang) * (lebar)

def luasHektarPersegi(panjang, lebar):
  result = int(panjang) * (lebar)

  return result / 10000

print(f" Luas dalam meter Persegi = {luasPersegi(inputPanjang, inputLebar)}")
print(f" Luas dalam hektar = {luasHektarPersegi(inputPanjang, inputLebar)} ")
```

3. Buatlah sebuah program yang dapat menghitung jumlah dari n bilangan pertama dari bilangan natural atau bilangan bulat positif. Misal jika user memberikan input n = 10 maka program harus menghitung jumlah dari 1 + 2+ 3+ … + 10

```python
n = input()

result = 0

for number in range(1, int(n)):
  result += number

print(result)
```

4. Buatlah program untuk menghitung luas segitiga dengan dua cara:

      a. Menghitung luas segitiga dengan rumus (alas x tinggi)/2

      b. Menghitung luas segitiga dengan rumus heron


```python
import math

print("1. luas segitiga")
print("2. luas segitiga dengan rumus heron")
menu = input("Pilih menu = ")

def luasSegitiga(alas, tinggi):
  return alas * tinggi / 2

def luasSegitigaHeron(a, b, c):
  s = (a + b + c) * 1/2
  return math.sqrt(s*(s - a) * (s - b) * (s - c))

if(int(menu) == 1):
  alas = input("masukkan nilai alas = ")
  tinggi = input("masukkan nilai tinggi = ")

  result = luasSegitiga(int(alas), int(tinggi))

  print(f"Luas segitiga dengan rumus (alas x tinggi)/2 = {result}")

elif(int(menu) == 2):
  
  a = input("masukkan nilai a = ")
  b = input("masukkan nilai b = ")
  c = input("masukkan nilai c = ")

  luas = luasSegitigaHeron(int(a), int(b), int(c))

  print(f"luas segitiga dengan rumus heron = {luas}")
else:

  print("maaf, tidak ada di pilihan")
```

Buatlah program yang membaca input magnitude gempa dari user dan program menunjukkan informasi deskripsi dari gempa tersebut. Contoh: Jika user memberi input 5.5 maka output dari program adalah “Moderate earthquake”

```python

def ritcherScale(magnitude):

  if(magnitude < 2.0):
    return "Micro earthquake"
  elif(magnitude >= 2.0 and magnitude <= 2.9):
    return "Minor earthquake"
  elif(magnitude >= 3.0 and magnitude <= 3.9):
    return "Minor earthquake"
  elif(magnitude >= 4.0 and magnitude <= 4.9):
    return "Light earthquake"
  elif(magnitude >= 5.0 and magnitude <= 5.9):
    return "Moderate earthquake"
  elif(magnitude >= 6.0 and magnitude <= 6.9):
    return "Strong earthquake"
  elif(magnitude >= 7.0 and magnitude <= 7.9):
    return "Major earthquake"
  elif(magnitude >= 8.0 and magnitude <= 8.9):
    return "Great earthquake"
  elif(magnitude >= 9.0 and magnitude <= 9.9):
    return "Great earthquake"
  elif(magnitude >= 10.0):
    return "Massive earthquake"

magnitude = input("masukkan nilai magnitude = ")

print(ritcherScale(float(magnitude)))
```
